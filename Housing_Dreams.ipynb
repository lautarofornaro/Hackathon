```python
# Carga de librerias

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import accuracy_score
```

```python
# Carga de CSV

Train_DF = pd.read_csv('.\Housing Dreams\house_train_raw.csv', encoding='latin-1')

Test_DF = pd.read_csv('.\Housing Dreams\houses_test_raw.csv', encoding='latin-1')
```

```python
# Primer vistazo a los datos

Train_DF.head()
```

```python
Test_DF.head()
```

```python
# Generamos una lista con las columnas teniendo en cuenta la correlacion con las variables
corr = pd.DataFrame()
corr['Columnas'] = Train_DF.corr()['SalePrice'].sort_values(ascending = False).index
corr['Valores'] = Train_DF.corr()['SalePrice'].sort_values(ascending = False).values
corr = corr[corr['Valores'] >= 0.60]
corr = corr['Columnas'].tolist()
corr.remove('SalePrice')
corr
```

```python
# Generamos una funcion para tratar los valores nulos
def Nulos(DF):
    for column in DF:
        DF[column].fillna(DF[column].mean(), inplace = True)
```

```python
# Features Train

features_train = Train_DF[corr]
Nulos(features_train)

X_train = features_train.values

# Features Test

features_test = Test_DF[corr]
Nulos(features_test)

X_test = features_test.values

# Target

target = Train_DF['SalePrice']

y_train = target.values
```

```python
# Creacion del Modelo "Arbol de Decision"

tree = DecisionTreeClassifier(max_depth = None)

# Fit del modelo

tree.fit(X_train, y_train)
```

```python
# Predict

y_train_pred = tree.predict(X_train)

y_test_pred = tree.predict(X_test)

# Visualizamos los scores

print("RMSLE:", mean_squared_log_error(y_train,y_train_pred))
print('accuracy_score:', accuracy_score(y_train_pred, y_train))
```

```python
# Generamos un DataFrame en el cual guardar las predicciones
y_preds = pd.DataFrame()
y_preds['preds'] = y_test_pred
y_preds.to_csv('y_preds.csv')
```

```python
# Creacion del Modelo "Random Forest"

forest = RandomForestClassifier(n_estimators= 1)

forest.fit(X_train, y_train)
```

```python
# Predict

y_train_pred = forest.predict(X_train)

y_test_pred = forest.predict(X_test)

# Visualizamos los scores

print("RMSLE:", mean_squared_log_error(y_train,y_train_pred))
print('accuracy_score:', accuracy_score(y_train_pred, y_train))
```

```python
# Generamos un DataFrame en el cual guardar las predicciones
y_preds = pd.DataFrame()
y_preds['preds'] = y_test_pred
y_preds.to_csv('y_preds_forest.csv')
```
